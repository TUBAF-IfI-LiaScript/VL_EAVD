# Matplotlib Beispiel

Einige Beispiele, welche die verschiedenen Diagrammtypen zeigen, die in den Übungen behandelt werden und prüfungsrelevant sind.

```python
from matplotlib import pyplot as plt
import numpy as np

# Generiere ein paar beliebige Beispielwerte
# x:     gemeinsame x Werte (Indizes)
# y:     Sinuskurve
# r1/r2: Zufallszahlen

x = np.linspace(0,4*np.pi,1001)
y = np.sin(x)
r1 = np.random.rand(1001)
r2 = np.random.randn(1001)
```

```python
plt.plot(x,y,'r')
```

```python
plt.scatter(x[::10],y[::10])
```

```python
plt.plot(x,r1)
```

```python
plt.hist(r1,bins=25)
```

```python
plt.hist(r2, bins=25)
```

```python
plt.scatter(x,r1,c=y,cmap="turbo")
```

```python
plt.scatter(x,r2)
```

```python
plt.boxplot((r1,r2))
```

# Diagrammformatierung und Export

Ziele:
- Legende
- Korrekte Achsenbeschriftung
- Ergebnis als Bilddatei speichern

```python
plt.plot(x, y, label="y = sin(x)", c="magenta")
plt.title("Sinuskurve (0 bis 4π)")
plt.legend()    # Legende anzeigen
plt.xlabel("x") # x-Achsenbeschriftung
plt.ylabel("y") # y-Achsenbeschriftung
plt.savefig("plot.png")#, dpi=200, bbox_inches="tight")
```

# Subplots

```python
fig,axs = plt.subplots(1,2, figsize=(14,6))
axs[0].scatter(x,r1, c='orange')
axs[0].set_title("Scatterplot für Gleichverteilung")
axs[1].scatter(x,r2, c='g')
axs[1].set_title("Scatterplot für Normalverteilung")
```
