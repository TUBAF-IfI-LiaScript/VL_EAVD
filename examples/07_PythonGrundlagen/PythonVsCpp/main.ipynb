# print-Statement in Python

print() dient der Ausgabe von Zeichenketten sowie beliebigen anderen Datentypen auf dem Bildschirm (oder auch dem Schreiben in Dateien).
Der print()-Funktion können beliebig viele Parameter übergeben werden.

```python
print("Hello", "World!")
```

```python
print("2**4 =", 2**4)
```
